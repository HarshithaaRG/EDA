```python
category_sub=df.groupby('category')['subscribers'].mean().sort_values(ascending=False)
category_sub
```

We can see that shows category channels have the highest average subscriber count

```python
category_sub.head().plot(kind='bar')
plt.title("Mean subscriber per category")
plt.xlabel("Category")
plt.ylabel("Mean subscibers")
plt.show()
```

```python
channel_type_counts=df['channel_type'].value_counts()
channel_type_counts.head().plot(kind='pie',autopct="%1.1f%%")
plt.title("Channel Type Distibution")
plt.show()
```

39.5% of the channels in youtube are entertainment channels
28.1% are music channels 
the top five are: entertainment, music, people, games, comedy

```python
plt.scatter(df['subscribers'],df["video views"])
plt.title("Scatter plot of subscribers vs video views")
plt.xlabel("Subscribers")
plt.ylabel("Video views")
plt.show()
```

```python
country_sub=df.groupby('Country')['subscribers'].sum().sort_values(ascending=False)
country_sub.head(10).plot(kind='bar')
plt.title("Total subscribers by country")
plt.xlabel("Country")
plt.ylabel("Total subscribers")
plt.show()
```

United states, india, brazil, uk and mexico are the countries with the most subscribers on youtube

```python
df.groupby("created_month")['subscribers'].sum().sort_values(ascending=False)
```

Sep 2023 had the most number of subscibers and oct had the least

```python
channel_upload=df.groupby("channel_type")["uploads"].sum().sort_values(ascending=False)
channel_upload.head().plot(kind='bar')
plt.title("Category wise upload")
plt.show()
```

As news channels keep giving updates from the world very frequently they have the highest uploads followed by entertainment. 

```python
sns.boxplot(data=df,x="channel_type",y="highest_monthly_earnings")
plt.title("Distribution of highest monthly earnings by channel type")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(['category','Country'])['subscribers'].sum().sort_values(ascending=False).unstack().head(5).plot(kind='bar',stacked=True)
plt.title("Subscriber breakdown by category and country")
plt.xlabel("category")
plt.ylabel("total subscribers")
plt.legend(title='country',bbox_to_anchor=(1.05,1),loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

Observation:
T-series channel has the highest number of subscribers
News channels upload more number of videos than other categories
United states has most number of overall subscribers
Show channels have the most number of average subscibers
Sep 2023 had the highest number of subscibers on youtube
39.5% of the channels on youtube is of entertainment category
