```python
# Exploratory Data Analysis on Black Friday Sale Dataset
# Dataset source : Kaggle https://www.kaggle.com/datasets/pranavuikey/black-friday-sales-eda/data
```

This dataset contains real-world transactional data collected during the Black friday sale (a major retail event). 
- The data describes customer demographic, product categories, and purchase behavior for thousands of shoppers.- - The objective of this EDA is to explore consumer behavior, spending patterns, and product demand across different segments such as age groups, gender, city categories, and product types.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("../dataset/black_friday_sale_data.csv")
df.head()
```

-Product_Category_2 has 31% of missing values (moderate %)
-Product_Category_3 has 69% of missing values(high %)
-No missing values are observed in other features

-The missing values in Product Category 2 and 3 indicate that certain products do not belong to secondary or tertiary product category. 

```python
df["Product_Category_2"]=df["Product_Category_2"].fillna(0)
df["Product_Category_3"]=df["Product_Category_3"].fillna(0)
df.isnull().sum()
```

-We have treated the missing values in the product category 2 and 3 by filling it with 0s meaning they don't have a secondary or tertiary category. 
- Now the data is consistent for analysis and visualisation

```python
df.nunique()
```

-User_ID and Product_ID have very high unique counts indicating that they are identifier columns

```python
sns.histplot(df["Purchase"],kde=True)
plt.title('Purchase Distribution')
plt.show()
```

The purchase distribution is a right skewed distribution indicating that most customers make purchases within 0-10,000 while  a small number of transactions involve high spending

```python
sns.boxplot(x=df['Purchase'])
plt.title("Outliers in Purchase amount")
plt.show()
```

- As the median is closer to the left side it is a right skewed distribution. Most customers make lower-value purchases
- A small number of customers have make very large purchases

```python
sns.countplot(data=df, x=df['Gender'])
plt.title('Gender Distibution')
plt.show()
```

-Indicates that the purchases by Male customers is very much greater than female customers


```python
sns.countplot(data=df,x=df['Age'])
plt.title("Age group distibution")
plt.show()
```

- This countplot indicates that the customers aged 36-35 have made the most number of purchases

```python
sns.barplot(data=df,x="Gender",y="Purchase")
plt.title("Gender vs Purchase")
plt.show()
```

- Male customers show a slightly higher average spending but when compared with the countplot we can observe that even though the number of purchasing females are less, their purchasing share/amount is more than males. 

```python
sns.barplot(data=df,x='Marital_Status',y='Purchase')
plt.title("Marital Status vs Purchase")
plt.show()
```

- In this analysis it shows that marital status does not have a significant effect in purchasing power.

```python
sns.lineplot(x='Age',y='Purchase',data=df,marker='o')
plt.title("Age group vs Purchase")
plt.show()
```

- So according to this people aged 51-55 have spent the most amount on purchases even though their count is relatively less than people aged 26-35

```python
df['City_Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

-Most of the customers belonged to City B

```python
top_products=df['Product_ID'].value_counts().head(10)
sns.barplot(x=top_products.index,y=top_products.values)
plt.xticks(rotation=45)
plt.title("Top 10 Products Sold in the Sale")
plt.show()
```

- Shows the top 10 Products sold during the Sale along with the count. 

```python
cols=['Purchase',"Product_Category_1","Product_Category_2","Product_Category_3"]
sns.heatmap(df[cols].corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Overall Conclusions:

- The purchase distribution tends to be right skewed indicating that people who make high transactions are less
- Females tend to be less in number but still spend more on an average than males. 
- Marital status does not have significant effect in purchase
- Product Category 1 the backbone while 2 and 3 are addons
- High value customers mostly sit within mid-age ranges.
-User_ID can be dropped as it has no significant effect in purchasing amount
